Tarefa: implementar a função “fit_k_means(pontos, parada, max_iter)”, com K fixo e igual a 3. Retorne os centroids finais.


```
import math
import random

# 1. Definir a quantidade K de clusters que queremos calcular
k = 3;


# 2. Sortear as coordenadas iniciais para cada um dos K medoids
medoid1 = [ int(pontos.iloc[[1]]["Casos"]), int(pontos.iloc[[1]]["Mortes"]) ];
medoid2 = [ int(pontos.iloc[[22]]["Casos"]), int(pontos.iloc[[22]]["Mortes"]) ];
medoid3 = [ int(pontos.iloc[[27]]["Casos"]), int(pontos.iloc[[27]]["Mortes"]) ];

custoTotal = 0;
novoCustoTotal = 0

cluster1 = [];
cluster2 = [];
cluster3 = [];

c1Total = 0;
c2Total = 0;
c3Total = 0;


# Inicio atribuirPonto
def atribuirPonto(ponto):

  #print('\nTeste = 2.0 \n');

  # Para utilizar as variáveis globais declaradas emcima do método
  global medoid1, medoid2, medoid3, custoTotal, novoCustoTotal, cluster1, cluster2, cluster3, c1Total, c2Total, c3Total;

  novoCustoTotal = custoTotal;

  c1 = ( medoid1[0] - int(ponto[0]) ) * ( medoid1[0] - int(ponto[0]) ) + ( medoid1[1] - int(ponto[1]) ) * ( medoid1[1] - int(ponto[1])); 

  c2 = ( medoid2[0] - int(ponto[0]) ) * ( medoid2[0] - int(ponto[0]) ) + ( medoid2[1] - int(ponto[1]) ) * ( medoid2[1] - int(ponto[1])); 

  c3 = ( medoid3[0] - int(ponto[0]) ) * ( medoid3[0] - int(ponto[0]) ) + ( medoid3[1] - int(ponto[1]) ) * ( medoid3[1] - int(ponto[1])); 

  #print('\nTeste = 2.1 \n');

  if(c1 < c2 and c1 < c3):
    cluster1.append(ponto);
    c1Total += c1;
  
  else:
  
    if(c2 < c1 and c2 < c3):
      cluster2.append(ponto);
      c2Total += c2;
    
    else:
    
      if(c3 < c1 and c3 < c2):
        cluster3.append(ponto);
        c3Total += c3;
  
  
  #print('\nTeste = 1.2 \n');

  if(len(cluster1) != 0):
    c1Total = c1Total/len(cluster1);
  
  if(len(cluster2) != 0):
    c2Total = c2Total/len(cluster2);

  if(len(cluster3) != 0):
    c3Total = c3Total/len(cluster3);

  novoCustoTotal = (c1Total + c2Total + c3Total)/k;

  if(custoTotal == 0):
    custoTotal = novoCustoTotal;

  else:
    if(novoCustoTotal < custoTotal):
      custoTotal = novoCustoTotal;

# Fim atribuirPonto


def fit_pam(pontos, alpha, max_iter):

  # Para utilizar as variáveis globais declaradas emcima do método
  global medoid1, medoid2, medoid3, custoTotal, novoCustoTotal, cluster1, cluster2, cluster3;
  
  m1 = 1;
  m2 = 22;
  m3 = 27;

  #Calcular o custo total MSE para os medoids acima:
  for x in range(1, len(pontos)):
    #print('\nTeste = 2 \n');
    ponto = [ int(pontos.iloc[[x]]["Casos"]), int(pontos.iloc[[x]]["Mortes"]) ];
    atribuirPonto(ponto);

  #print('\nTeste = 3 \n');

  #Sortear novos pontos como Medoids;
  for y in range(alpha):
    cluster1 = [];
    cluster2 = [];
    cluster3 = [];

    c1Total = 0;
    c2Total = 0;
    c3Total = 0;

    novo = 0;
    pontoAntigo = 0;

    z = random.randint(1,len(pontos))

    if(z != m1 and z != m2 and z != m3):

      #print('\nTeste = 4 \n');
      
      ponto = [int(pontos.iloc[[z]]["Casos"]), int(pontos.iloc[[z]]["Mortes"]) ];

      atribuirPonto(ponto);

      #print('\nTeste = 5 \n');

      if(len(cluster1) > 0):
        pontoAntigo = m1;
        m1 = z;
        medoid1 = [ int(pontos.iloc[[z]]["Casos"]), int(pontos.iloc[[z]]["Mortes"]) ];
        novo = 1;

      else:
        #print('\nTeste = 6 \n');

        if(len(cluster2) > 0):
          pontoAntigo = m2;
          m2 = z;
          medoid2 = [ int(pontos.iloc[[z]]["Casos"]), int(pontos.iloc[[z]]["Mortes"]) ];
          novo = 2;
        
        else:
          #print('\nTeste = 7 \n');

          if(len(cluster3) > 0):
            pontoAntigo = m3;
            m3 = z;
            medoid3 = [ int(pontos.iloc[[z]]["Casos"]), int(pontos.iloc[[z]]["Mortes"]) ];
            novo = 3;

      cluster1 = [];
      cluster2 = [];
      cluster3 = [];

      #print('\nTeste = 8 \n');

      #atribuir Todos Pontos para o novo medoid:
      for x in range(len(pontos)):
        ponto = [ int(pontos.iloc[[x]]["Casos"]), int(pontos.iloc[[x]]["Mortes"]) ];
        atribuirPonto(ponto);

        #print('\nTeste = 9 \n');
      
      if(novoCustoTotal > custoTotal):
        if(novo == 1):
          medoid1 = [ int(pontos.iloc[[pontoAntigo]]["Casos"]), int(pontos.iloc[[pontoAntigo]]["Mortes"]) ];
        
        else:
          if(novo == 2):
            medoid2 = [ int(pontos.iloc[[pontoAntigo]]["Casos"]), int(pontos.iloc[[pontoAntigo]]["Mortes"]) ];

          else:
            if(novo == 3):
              medoid3 = [ int(pontos.iloc[[pontoAntigo]]["Casos"]), int(pontos.iloc[[pontoAntigo]]["Mortes"]) ];


  return medoid1, medoid2, medoid3;
```
